```python
import lightgbm as lgb
from sklearn.datasets import fetch_covtype
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split

# データセットの読み込み
covertype = fetch_ucirepo(id=31)

X= covertype.data.features
y = covertype.data.targets
y -= 1
```

```python
# StratifiedKFoldによるデータの分割とモデルの訓練・評価
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []

for fold_num, (train, valid) in enumerate(skf.split(X, y)):
    print(f"Fold {fold_num+1}:")

    # 訓練データとラベル
    train_data, train_label = X.iloc[train], y.iloc[train]
    
    # テストデータとラベル
    valid_data, valid_label = X.iloc[valid], y.iloc[valid]
    
    # データの分割
    X_train, X_test, y_train, y_test = train_test_split(train_data, train_label, test_size=0.2, random_state=42)

    # LightGBM用のデータセットに変換
    train_data = lgb.Dataset(X_train, label=y_train)
    test_data = lgb.Dataset(X_test, label=y_test)

    ################################################LightGBMモデルの部分はchatgpt使用############################################################
    
    # LightGBMのハイパーパラメータの設定
    params = {'objective': 'multiclass',  # 損失関数(2クラス分類の場合'binary'、多クラス分類の場合'multiclass'、回帰の場合'regression')
             'metric': 'multi_logloss',  # 評価指標(2クラス分類の場合'binary_logloss'等、多クラス分類の場合'multi_logloss'、回帰の場合'rmse'等)
             'random_state': 42,  # 乱数シード
             'boosting_type': 'gbdt',  # boosting_type
             'num_class': '7',  # クラスの数を指定する
             'verbose': -1  # これを指定しないと`No further splits with positive gain, best gain: -inf`というWarningが表示される
             }

    verbose_eval = 1  # この数字を1にすると学習時のスコア推移がコマンドライン表示される

    # LightGBMモデルの学習
    model = lgb.train(params, train_data, # early_stoppingの評価用データ
                    num_boost_round=1000,  # 最大学習サイクル数。early_stopping使用時は大きな値を入力
                    valid_sets=test_data,
                    callbacks=[lgb.early_stopping(stopping_rounds=10, 
                                    verbose=True), # early_stopping用コールバック関数
                               lgb.log_evaluation(verbose_eval)] # コマンドライン出力用コールバック関数
                    )
    
        ###############################################LightGBMモデルの部分はchatgpt使用############################################################

    # テストデータでの予測
    y_pred = model.predict(valid_data)
    # 最も確率の高いクラスを予測値として取得
    y_pred_class = np.argmax(y_pred, axis=1)

    #y_test += 1 # テストデータのクラスの値を調整
    # 精度の評価
    accuracy = accuracy_score(valid_label, y_pred_class)
    print('Accuracy:', accuracy)
    accuracies.append(accuracy)

    # 混同行列を計算
    conf_matrix = confusion_matrix(valid_label, y_pred_class)

    # 混同行列をヒートマップとしてプロット
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Confusion Matrix - Fold {}'.format(fold_num + 1))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
    
    plt.figure(figsize=(30,6))
    lgb.plot_importance(model)
    plt.show()
    
# 精度の平均を表示
print("Accuracy:", accuracies)
print("Mean Accuracy:", np.mean(accuracies))
```
